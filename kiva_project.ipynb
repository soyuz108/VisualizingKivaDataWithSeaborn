## Visualizing Loans Awarded by Kiva

In this project we'll visualize insights using a dataset from <a href = "https://www.kaggle.com/fkosmowski/kivadhsv1" target = "_blank">Kaggle</a>. The dataset contains information about loans awarded by the non-profit <a href = "https://www.kiva.org/" target = "_blank">Kiva</a>. Using Seaborn, we'll explore the average loan amount by country using aggregated bar charts. We'll also visualize the distribution of loan amounts by project type and gender using box plots and violin plots.

Each entry (row) in the dataset represents a loan that Kiva awarded to a particular project. The `loan_amount` column shows the amount (in U.S. dollars) awarded to the project. The `activity` column has the category type that the project falls under. The `country` column is the country where the project is located. The `gender` column represents the gender of the primary person who applied for the loan. 

### Getting ready

```python
# Import libraries
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
```

```python
# Load the raw data and inspect it
df = pd.read_csv("kiva_data.csv")
df.head(10)
```

### Visualizing bar plots

```python
# Create the figure to visualize loans by country
f, ax = plt.subplots(figsize=(8, 6))

# Plot the data
sns.barplot(data=df, x="country", y ="loan_amount")

# Edit the y-axis ticks
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

```python
# Creates the figure to visualize loans by country, nesting by gender
f, ax = plt.subplots(figsize=(8, 6))

sns.barplot(data=df, x="country", y="loan_amount", hue="gender")

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

```python
# Set style
sns.set_style("whitegrid")

# Set color palette
sns.set_palette("pastel")

# Create a figure and axes
plt.figure(figsize=(8, 6))

# Add a title
plt.title("Average Loan Amounts Given by Kiva, by country, by gender (in $ USD)")

# Use Seaborn to create the bar plot
sns.barplot(data=df, x="country", y="loan_amount", hue="gender")
plt.show()
```

```python
# Set color palette
sns.set_palette("muted")

# Create a figure and axes to compare the distribution of loans by country
plt.figure(figsize=(8, 6))

# Add a title
plt.title("Average Size of Loans by Country")

# Use Seaborn to create the bar plot
sns.boxplot(data=df, x="country", y="loan_amount")
plt.show()
```

Instead of visualizing the loan amount by country, we now want to use plot the loan amount by activity.

```python
# Set color palette
sns.set_palette("bright")

# Create a figure and axes to visualize loans by activity
plt.figure(figsize=(8, 6))

# Add a title
plt.title("Average Size of Loans by Activity")

# Use Seaborn to create the bar plot
sns.boxplot(data=df, x="activity", y="loan_amount")
plt.show()
```

### Visualizing violin plots

```python
# Set color palette
sns.set_palette("colorblind")

# Create a figure and axes to visualize the distribution of loan amounts by activity
plt.figure(figsize=(8, 6))

# Add a title
plt.title("Distribution of Loans by Activity")

# Use Seaborn to create the bar plot
sns.violinplot(data=df, x="activity", y="loan_amount")
plt.show()
```

```python
# Set color palette
sns.set_palette("dark")

# Create a figure and axes to visualize the distribution of loan amounts by country
plt.figure(figsize=(8, 6))

# Add a title
plt.title("Distribution of Loans by Activity")

# Use Seaborn to create the bar plot
sns.violinplot(data=df, x="country", y="loan_amount")
plt.show()
```

```python
# Change a style
sns.set_palette("deep")

# Create a figure and axes to visualize the distribution of loan amounts by country, split by gender
plt.figure(figsize=(8, 6))

# Add a title
plt.title("Distribution of Loans by Activity (split by Gender)")

sns.violinplot(data=df, x="country", y = "loan_amount", hue="gender", split=True)
plt.show()
```
